```python
def calc(operation):
    v = 0
    if operation.find('плюс') != -1:
        operation = operation.replace(' плюс ', ' ')
        for i in range(len(operation.split())):
            v += dict_operation[operation.split()[i]]
    elif operation.find('минус') != -1:
        operation = operation.replace(' минус ', ' * ')
        k = 6
        for i in range(len(operation.split(" "))):
            if i < k:
                v += dict_operation[operation.split(' ')[i]]
                if operation.split()[i] == '*':
                    k = i
            else:
                v -= dict_operation[operation.split()[i]]
    return v
 dict_operation = {'один': 1, 
                  'два': 2,
                  'три': 3,
                  'четыре': 4,
                  'пять': 5,
                  'шесть': 6,
                  'семь': 7,
                  'восемь': 8,
                  'девять': 9,
                  'десять': 10,
                  'одиннадцать': 11,
                  'двенадцать': 12,
                  'тринадцать': 13,
                  'четырнадцать': 14,
                  'пятнадцать': 15,
                  'шестнадцать': 16,
                  'семнадцать': 17,
                  'восемнадцать': 18,
                  'девятнадцать': 19,
                  'двадцать': 20,
                  'тридцать': 30,
                  'сорок': 40,
                  'пятьдесят': 50,
                  'шестьдесят': 60,
                  'семьдесят': 70,
                  'восемьдесят': 80,
                  'девяносто': 90,
                  'сто': 100,
                  '*': 0,
                  'двести': 200}
p = calc(input('Введите математическое выражение '))
k = ''
if p <= 20:
    for i, j in dict_operation.items():
        if j == p:
            print(i)
else:
    if p >= 100:
        k += 'сто' + ' '
        p -= 100
    if 20 < p % 100 < 100:
        for i, j in dict_operation.items():
            if j == (p // 10) * 10:
                k += i + " "
                p = p - (p // 10) * 10
                break
         
    if p % 100 <= 20:
        for i, j in dict_operation.items():
            if j == p:
                k += i
                break
                
print(k)
    
```
