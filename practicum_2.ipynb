```python
def calc(operation: str) -> int or float:# главная функция, вычисляющая алгебраические операции.
    if operation.find('плюс') != -1:
        v, v_1 = float_or_integer_processing(operation.replace(' плюс ', ' ноль '))
        return v + v_1
    elif operation.find('минус') != -1:
        v, v_1 = float_or_integer_processing(operation.replace(' минус ', ' ноль '))
        return v - v_1
    elif operation.find('разделить') != -1:
        v, v_1 = float_or_integer_processing(operation.replace(' разделить на ', ' ноль '))
        return v / v_1


def integer_thousand(integer: str) -> int: # функция вычисляющая тысячныю часть числа
    s = 0
    for i in integer.split(' '):
        s += dict_operation[i]
    return s


def float_or_integer_processing(operation: str) -> int or float: # функция преобразующая строковой тип данных в числовой.
    v = 0      # целая часть 1-ого числа
    v_1 = 0    # целая часть 2-ого числа
    v_dr = 0   # дробная часть 1-ого числа
    v_dr1 = 0  # дробная часть 2-ого числа
    k, l, w, flag = 10 ** 10, 10 ** 10, 10 ** 10, True
    operation.replace('тысяча', 'тысяч')
    v_thousand = 0
    v_thousand_1 = 0
    if operation.find('тысяч') != -1 and operation.find('тысячн') != operation.find('тысяч'):
        v_thousand = integer_thousand(operation[0:operation.find('тысяч')-1])
        operation = operation[operation.find('тысяч')+7:]
    if operation.rfind('тысяч') != -1 and operation.rfind('тысячн') != operation.rfind('тысяч'):
        v_thousand_1 = integer_thousand(operation[operation.rfind('ноль')+5:operation.rfind('тысяч')-1])
        operation = operation[:operation.rfind('ноль')+5] + operation[operation.rfind('тысяч')+6:]
    for i in range(len(operation.split(" "))):
        if i < k and i < l:
            v += dict_operation[operation.split(' ')[i]]
            if operation.split()[i] == 'и':
                l = i
            elif operation.split()[i] == 'ноль':
                k = i
        elif i > l and i < k and operation.split()[i] not in list_drobi:
            v_dr += dict_operation[operation.split(' ')[i]]
            if operation.split()[i] == 'ноль':
                k = i
        elif operation.split()[i] in list_drobi and flag:
            v_dr /= dict_operation_drobi[operation.split()[i]]
            v += v_dr
        else:
            flag = False
            if w > i:
                v_1 += dict_operation[operation.split()[i]]
                if operation.split()[i] == 'и':
                    w = i
            elif w < i and operation.split()[i] not in list_drobi:
                v_dr1 += dict_operation[operation.split(' ')[i]]
            elif operation.split()[i] in list_drobi:
                v_dr1 /= dict_operation_drobi[operation.split()[i]]
                v_1 += v_dr1
    v = v_thousand * 1000 + v
    v_1 = v_thousand_1 * 1000 + v_1
    return v, v_1


def translatep_integer(num: int) -> str: # преобразует целочисленный тип в строковый
    k = ''
    if 999 >= num >= 100:
        k += ''.join([i for i, j in dict_operation_integer.items() if j == (num // 100) * 100]) + ' '
        num -= (num // 100) * 100
    if 20 < num % 100 < 100:
        for i, j in dict_operation_integer.items():
            if j == (num // 10) * 10:
                k += i + " "
                num -= (num // 10) * 10
                break
    if 1 <= num % 100 <= 20:
        k += ''.join([i for i, j in dict_operation_integer.items() if j == num])
    if k == '' and num == 0:
        return 'ноль'
    return k

def translatep_float(num: float) -> str: # преобразует дробный тип в строковый
    k = ''
    if num >= 100:
        k += ''.join([i for i, j in dict_operation_float.items() if j == (num // 100) * 100]) + ' '
        num -= (num // 100) * 100
    if 20 < num % 100 < 100:
        for i, j in dict_operation_float.items():
            if j == (num // 10) * 10:
                k += i + " "
                num -= (num // 10) * 10
                break
    if 1 <= num % 100 <= 20:
        k += ''.join([i for i, j in dict_operation_float.items() if j == num])
    if k == '' and num == 0:
        return 'ноль'
    return k


dict_operation_drobi = {'десятая': 10,
                        'десятых': 10,
                        'сотая': 100,
                        'сотых': 100,
                        'тысячная': 1000,
                        'тысячных': 1000,
                        'десятитысячная': 10000,
                        'десятитысячных': 10000,
                        'статысячная': 100000,
                        'статысячных': 100000,
                        'миллионная': 1000000,
                        'миллионных': 1000000}

list_drobi = ['десятых', 'десятая', 'сотых', 'сотая', 'тысячных', 'тысячная', 'десятитысячная', 'десятитысячных', 'статысячная', 'статысячных', 'миллионная', 'миллионных']


dict_operation = {'ноль': 0,
                  'один': 1,
                  'одна': 1,
                  'два': 2,
                  'две': 2,
                  'три': 3,
                  'четыре': 4,
                  'пять': 5,
                  'шесть': 6,
                  'семь': 7,
                  'восемь': 8,
                  'девять': 9,
                  'десять': 10,
                  'одиннадцать': 11,
                  'двенадцать': 12,
                  'тринадцать': 13,
                  'четырнадцать': 14,
                  'пятнадцать': 15,
                  'шестнадцать': 16,
                  'семнадцать': 17,
                  'восемнадцать': 18,
                  'девятнадцать': 19,
                  'двадцать': 20,
                  'тридцать': 30,
                  'сорок': 40,
                  'пятьдесят': 50,
                  'шестьдесят': 60,
                  'семьдесят': 70,
                  'восемьдесят': 80,
                  'девяносто': 90,
                  'сто': 100,
                  'двести': 200,
                  'триста': 300,
                  'четыреста': 400,
                  'пятьсот': 500,
                  'шестьсот': 600,
                  'семьсот': 700,
                  'восемьсот': 800,
                  'девятьсот': 900,
                  'тысяча': 1000,
                  'и': 0}


dict_operation_float = {i: k for i, k in dict_operation.items() if i != 'один' and i != 'два'}


dict_operation_integer = {i: k for i, k in dict_operation.items() if i != 'одна' and i != 'две'}

flag_1 = False
p = calc(input('Введите строку для математических преобразований: '))
if type(p) == int:
    if p >= 1000:
        print(translatep_integer(p//1000) + ' тысяч ' + translatep_integer(p % 1000))
    else:
        print(translatep_integer(p))
else:
    l_3 = ''
    l_2 = ''
    p = str(round(p, 20)).split('.')
    l_1 = translatep_integer(int(p[0]))
    for i in range(len(p[1])):
        for period in range(1, 5):
            if p[1][i:i+period] == p[1][i+period:i+period*2] == p[1][i+period*2:i+period*3]:
                l_3 = p[1][i:i+period]
                flag_1 = True
                break
        if flag_1 or i == 6:
            break
        l_2 += p[1][i]
    if l_2:
        l_2 = translatep_drobi_part(l_2)
        if l_3:
            if l_3[0] == '0':
                l_3 = translatep_integer(int(l_3[1:]))
                print(l_1 + ' и ' + l_2 + ' и ноль ' + l_3 + ' в периоде')
            elif l_3[:2] == '00':
                l_3 = translatep_integer(int(l_3[1:]))
                print(l_1 + ' и ' + l_2 + ' и ноль ноль ' + l_3 + ' в периоде')
            elif l_3[:3] == '000':
                l_3 = translatep_integer(int(l_3[1:]))
                print(l_1 + ' и ' + l_2 + ' и ноль ноль ноль ' + l_3 + ' в периоде')
            else:
                l_3 = translatep_integer(int(l_3))
                print(l_1 + ' и ' + l_2 + ' и ' + l_3 + ' в периоде')
        else:
            print(l_1 + ' и ' + l_2)
    else:
        if l_3[0] == '0':
            l_3 = translatep_integer(int(l_3[1:]))
            print(l_1 + ' и ноль ' + l_3 + ' в периоде')
        elif l_3[:2] == '00':
            l_3 = translatep_integer(int(l_3[1:]))
            print(l_1 + ' и ноль ноль ' + l_3 + ' в периоде')
        elif l_3[:3] == '000':
            l_3 = translatep_integer(int(l_3[1:]))
            print(l_1 + ' и ноль ноль ноль ' + l_3 + ' в периоде')
        else:
            l_3 = translatep_integer(int(l_3))
            print(l_1 + ' и ' + l_3 + ' в периоде')
```
